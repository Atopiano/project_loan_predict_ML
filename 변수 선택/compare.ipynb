# 변수 선택 전과 후 비교

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

data = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC-AUC'],
    'All Features': [0.9616315, 0.9370629, 0.9897379, 0.9626804, 0.9935473],
    'Selected Features': [0.9736257, 0.9529316, 0.9964704, 0.9742148, 0.9974812]
}

df = pd.DataFrame(data, index = ["Acc", "Pre", "Re","F1", "AUC"])
df.reset_index
df = df.set_index("Metric")
df
```

```python
import matplotlib.pyplot as plt

# x-axis labels
metrics = df.index

# y-axis data
all_features_data = df['All Features']
selected_features_data = df['Selected Features']

# create a figure and axis object
fig, ax = plt.subplots()

# plot the data with customized line colors
ax.plot(metrics, all_features_data, marker='o', label='All Features', color='purple')
ax.plot(metrics, selected_features_data, marker='o', label='Selected Features', color='#8977AD')

# set axis labels and title
ax.set_xlabel('Metric')
ax.set_ylabel('Score')
ax.set_title('Comparison of All Features and Selected Features')

# set legend
ax.legend()

# display the plot
plt.show()
```
