```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'Model': ['CatBoost_ALL','Decision Tree_ALL', 'LGBM_ALL', 'RandomForest_ALL', 'XGBoost_ALL'],
        'Accuracy': [0.9589516,0.9329368,0.9598340,0.9646021, 0.9462056],
        'Precision': [0.9334527, 0.9151363, 0.9343665, 0.9471532, 0.9165294],
        'Recall': [0.9883653, 0.9543761, 0.9891496, 0.9841156, 0.9818289 ],
        'F1 Score': [0.9601245, 0.9343444,  0.9609779, 0.9652807,  0.9480560],
        'ROC-AUC': [0.9928379 , 0.9329368 ,0.9928593 ,0.9946687 ,0.9863311]
        }

df = pd.DataFrame(data)
df.set_index('Model', inplace=True)
plt.figure(figsize=(10, 5))
sns.heatmap(data = df, annot=True,fmt=".6f",cmap='Purples', xticklabels=df.columns, yticklabels=df.index)
plt.show()
```

```python
accuracy_rank = pd.DataFrame(index=df.Accuracy.index)
for metric in df.columns:
    rank = df[metric].rank(method='min', ascending=False)
    accuracy_rank = accuracy_rank.join(rank, rsuffix=' Rank')
accuracy_rank['Total Rank'] = accuracy_rank.sum(axis=1)
accuracy_rank['Rank'] = accuracy_rank['Total Rank'] .rank(method='dense', ascending=False)

```

```python
accuracy_rank.sort_values(by = "Rank")
accuracy_rank[["Rank"]].sort_values(by = "Rank")
```
